## 3. Estado del Arte

Revisamos cuatro trabajos que abordaron problemas similares de predicción de churn en servicios de suscripción. La idea es entender qué hicieron, qué funcionó y qué no, para no reinventar la rueda.

### Conclusiones de papers revisados

**Devan (2023)** usa 5 clasificadores clásicos sobre datos de encuestas de usuarios OTT. El Gradient Boosting sale ganador en AUROC. El problema: usan Train/Test split sin validación cruzada k-fold, así que la estimación es un poco frágil. Pero el punto importante es que en streaming de video (similar a música), Gradient Boosting tiende a funcionar bien.

**Zhang (2023)** hace un análisis más exhaustivo: 9 algoritmos sobre 21 características. La metodología es sistemática: la misma partición Train/Test para todos, métricas comparables. Logística logra 79.6%. Lo bueno es que enfatiza la preparación de datos y la evaluación bajo las mismas condiciones.

**Ekweli (2025)** es el más relevante para nosotros. Trabaja con desbalance significativo de clases (como nuestro dataset: 70% vs 30%). Usa LR, RF y XGBoost. Resultado: Random Forest y XGBoost alcanzan 0.79 accuracy y 0.83 AUC. Lo importante: demuestra que estos modelos son robustos ante desbalance si se aplican bien (normalización, one-hot-encoding, manejo de class weights).

**Mezentseva et al. (2021)** aporta metodología interesante: no solo reporta accuracy, sino que analiza curvas de validación (train vs. test) para entender sobreajuste. Random Forest logra 0.87 accuracy. Descubren que SVM y Decisión Trees sobreajustan. Esto nos sugiere que RF es más estable, pero tendremos que comprobarlo en este caso.

### Señales

1. **Random Forest y XGBoost ganan en casi todos los contextos.** No es coincidencia. Estos modelos manejan bien multicolinealidad, desbalance y relaciones no lineales.

2. **El desbalance es un factor crítico.** Ekweli (2025) lo trata explícitamente. Nuestro dataset tiene 70% no-churn y 30% churn, así que necesitamos técnicas de balanceo.

4. **Validación cruzada es mejor que Train/Test único.** La mayoría de trabajos usan Train/Test, pero eso deja mucha varianza. Validación cruzada (5-fold) da más confianza.

5. **La multicolinealidad no aparece explícitamente en estos trabajos, pero es problema conocido.** Nuestro EDA encontró correlación fuerte entre `ads_listened_per_week` y `offline_listening`. Hay que manejarlo.
